```python
## Cuál es el género con más revies?
df = ratings.merge(movies, on="movieId")

df_genres = df.assign(genre=df["genres"].str.split('|')).explode("genre")#Separar géneros 

reviews_por_genero = df_genres.groupby("genre")["rating"].count().sort_values(ascending=False)#Contar reviews por género

print(reviews_por_genero.head(5))#Dejamos solo un top 5
```

```python
## Cuáles fueron los géneros mejor rankeados en promedio en el año 2005?
ratings["date"] = pd.to_datetime(ratings["timestamp"], unit="s")

# Filtrar solo las reviews de 2005
r_2005 = ratings[ratings["date"].dt.year == 2005]

# Unir con movies
df = r_2005.merge(movies, on="movieId")

# Separar géneros
df_genres = df.assign(genre=df["genres"].str.split('|')).explode("genre")

# Calcular promedio de rating por género en 2005
p_2005 = df_genres.groupby("genre")["rating"].mean().sort_values(ascending=False)

print(p_2005.head(5))
```

```python
## Cuál es el top 5 de películas mejor rankeadas en promedio?
user_counts = ratings.groupby("userId")["rating"].count().sort_values(ascending=False)
top_user = user_counts.index[0]
print("Usuario con más críticas:", top_user)

# === 2. Género favorito del usuario ===
# Filtrar ratings de ese usuario
user_ratings = ratings[ratings["userId"] == top_user]

# Unir con movies para ver géneros
df_user = user_ratings.merge(movies, on="movieId")

# Separar géneros (porque están en formato "Action|Adventure|Drama")
df_user_genres = df_user.assign(genre=df_user["genres"].str.split('|')).explode("genre")

# Contar cuántas críticas hizo por género
fav_genre = df_user_genres["genre"].value_counts().idxmax()
print("Género favorito del usuario:", fav_genre)
```
